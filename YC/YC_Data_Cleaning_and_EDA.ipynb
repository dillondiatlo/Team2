```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify = pd.read_csv('./Data/SpotifyFeatures.csv')
tracks = pd.read_csv('./Data/tracks.csv')
```

## EDA

```python
plt.figure(figsize = (15,8))
spotify['genre'].value_counts().plot(kind='bar')
plt.xlabel('Count')
plt.xticks(rotation = 45)
plt.ylabel('Genre')
plt.title('Genre Distribution')
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.displot(spotify['popularity'],bins=50,kde = True)
plt.title('Popularity histgram')
plt.margins(x=0.05, y=0.05)  # Decrease white edge
plt.tight_layout() 
plt.show()
```

```python
pop_0 = spotify[spotify['popularity']<=20]
```

```python
pop_0.columns
```

```python
sns.heatmap(pop_0.corr(numeric_only=True)[['popularity']].sort_values(by='popularity',ascending = False)
            ,annot=True,cmap='Paired', fmt=".2f", linewidths=.5, vmin=-1, vmax=1);
```

```python
plt.figure(figsize = (12,8))
pop_0['genre'].value_counts().plot(kind='bar')
plt.xlabel('Count')
plt.xticks(rotation = 45)
plt.ylabel('Genre')
plt.title('Genre Distribution')
```

```python
sns.heatmap(spotify.corr(numeric_only=True)[['popularity']].sort_values(by='popularity',ascending =False),annot=True,cmap='coolwarm')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(spotify.corr(numeric_only=True),annot=True,cmap='Paired', fmt=".2f", linewidths=.5, vmin=-1, vmax=1)
plt.xticks(rotation = 34);
```
