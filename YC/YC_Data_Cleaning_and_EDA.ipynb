```python
genre_dum = pd.get_dummies(data=spotify[['track_id', 'genre']], columns=['genre'], dtype=int)
genre_dum_2 = genre_dum.groupby('track_id').sum()
genre_dum_2.shape
```

```python
genre_dum_2.columns
```

```python
spotify_pre_merge = spotify.drop(columns = ['genre'],axis = 1).drop_duplicates()
```

```python
spotify_genre_undup = pd.merge(
    left = spotify_pre_merge, 
    right = genre_dum_2,
    left_on = 'track_id',
    right_on = genre_dum_2.index,
    how = 'right'
    
)
spotify_genre_undup.sort_values('popularity', ascending = False).head(10)
```

```python
spotify_genre_undup.shape
```

```python
plt.figure(figsize = (12,8))
spotify_genre_undup['key'].value_counts().plot(kind='bar', color = 'darkorchid')
plt.xlabel('Count')
plt.xticks(rotation = 45)
plt.ylabel('Frequency')
plt.title('Key Distribution among all Popularity')
```
